##Penerapan functional programming (lambda, recursion, first class)

lambda

```python
# Mengubah kolom 'streams' ke tipe data numerik
musik['streams'] = pd.to_numeric(musik['streams'], errors='coerce')  # Mengubah ke float, NaN untuk nilai yang tidak valid
# Menggunakan lambda function untuk mengkategorikan berdasarkan jumlah streams
musik['Stream Category'] = musik['streams'].apply(lambda x: 'High' if x > 1e9 else ('Medium' if x > 5e8 else 'Low'))
musik.head()
```

```python
# Mengatur tampilan pandas untuk melihat semua kolom
pd.set_option('display.max_columns', None)
print(musik.head())
```

recursion

```python
musik.dropna(subset=['streams'], inplace=True)
# Fungsi rekursif untuk menjumlahkan streams
def recursive_sum(data, idx=0):
    if idx >= len(data):  # Periksa jika index melebihi panjang data
        return 0
    return data[idx] + recursive_sum(data, idx + 1)
# Memanggil fungsi rekursif untuk menghitung total streams
total_streams = recursive_sum(musik['streams'].tolist())  # Mengubah kolom 'streams' menjadi list
print(f"Total Streams: {total_streams}")
```

first class function

```python
# fungsi filter_high_streams yang bisa diperlakukan sebagai objek
def filter_high_streams(musik, threshold):
    return musik[musik['streams'] > threshold] #nilai ambang batas
# Memanggil fungsi filter dengan menggunakan first class function
high_stream_songs = filter_high_streams(musik, 1e9) #1 miliar
high_stream_songs.head()
```

##Penerapan numpy (statistic, join, splitting, matriks)

statistic

```python
import pandas as pd
import numpy as np
mean_streams = np.mean(musik['streams'])  # Menghitung rata-rata streams
median_streams = np.median(musik['streams'])  # Menghitung median streams
std_streams = np.std(musik['streams'])  # Menghitung standar deviasi
print(f"Mean: {mean_streams}, Median: {median_streams}, Std: {std_streams}")
```

join

```python
# Menghitung total streams per artis
artist_streams = musik.groupby('artist(s)_name')['streams'].sum().reset_index()  # Mengelompokkan berdasarkan artis
print("\nTotal Streams per Artist:")
print(artist_streams.sort_values(by='streams', ascending=False).head(10))
```

splitting

```python
# Membagi array streams menjadi 3 bagian menggunakan numpy
stream_chunks = np.array_split(artist_streams['streams'], 3)  # Membagi streams menjadi 3 bagian
# Menampilkan hasil split
print("\nHasil Pembagian Streams:")
for i, chunk in enumerate(stream_chunks):
    print(f"Chunk {i+1}:")
    print(chunk)
```

matrix

```python
# Membuat matriks sederhana dari kolom numerik
matrix = np.array([artist_streams['streams'].values, np.arange(len(artist_streams))])
# Melakukan operasi pada matriks (misalnya mengalikan dengan skalar)
matrix_scaled = matrix * 1.5
print("\nMatriks yang Diperbesar:")
print(matrix_scaled)
```

##Penerapan visualization (matplotlib, seaborn)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mengelompokkan data berdasarkan artis dan menghitung total streams per artis
artist_streams = musik.groupby('artist(s)_name')['streams'].sum().reset_index()

# Mengurutkan artis berdasarkan total streams, kemudian menampilkan 10 artis teratas
top_10_artists = artist_streams.sort_values(by='streams', ascending=False).head(10)

# Membuat bar plot total streams per artis untuk 10 artis teratas
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_artists, x='artist(s)_name', y='streams')
plt.xticks(rotation=45)
plt.title('Top 10 Artists by Total Streams')
plt.xlabel('Artist Name')
plt.ylabel('Total Streams')
plt.show()

```

```python
import warnings
warnings.filterwarnings("ignore")
musik_2021 = musik[musik['released_year'] == 2021]
# Mengurutkan berdasarkan jumlah streams dan memilih 5 lagu teratas
top_5_lagu = musik_2021.nlargest(5, 'streams')
# Membuat bar plot untuk 5 lagu paling terkenal di tahun 2021
plt.figure(figsize=(12, 6))
sns.barplot(data=top_5_lagu, x='track_name', y='streams', palette='viridis')
plt.title('5 Lagu Paling Terkenal di 2021')
plt.xlabel('Nama Lagu')
plt.ylabel('Jumlah Streams')
plt.xticks(rotation=45)
plt.show()
```
